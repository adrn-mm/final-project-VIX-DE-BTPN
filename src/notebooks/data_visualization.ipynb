# Import Datasets

```python
df_status = pd.read_csv('../../data/interim/segment_status.csv')
df_gender = pd.read_csv('../../data/interim/segment_gender.csv')
df_card = pd.read_csv('../../data/interim/segment_card.csv')
df_dependent = pd.read_csv('../../data/interim/segment_dependent.csv')
df_education = pd.read_csv('../../data/interim/segment_education.csv')
df_income = pd.read_csv('../../data/interim/segment_income.csv')
df_marital = pd.read_csv('../../data/interim/segment_marital.csv')
df_age = pd.read_csv('../../data/interim/segment_age.csv') 
```

# Feature Engineering

```python
df_gender = df_gender.pivot(index='Gender', columns='Status', values='Total')
df_card = df_card[['CardCategory', 'Total']].groupby(['CardCategory'], as_index=False).sum()

# manipulate df_education
df_education = df_education[['Education', 'Total']].groupby(['Education'], as_index=False).sum()
df_education.set_index('Education', inplace=True)
df_education.sort_values(by='Total', ascending=True, inplace=True)
df_education.drop('Unknown', inplace=True)

# manipulate df_dependent
df_dependent = df_dependent[['DependentCount', 'Total']].groupby(['DependentCount'], as_index=False).sum()
df_dependent.set_index('DependentCount', inplace=True)
df_dependent.sort_values(by='Total', ascending=True, inplace=True)

# manipulate df_income
df_income = df_income[['IncomeCategory', 'Total']].groupby(['IncomeCategory'], as_index=False).sum()
df_income.set_index('IncomeCategory', inplace=True)
df_income.sort_values(by='Total', ascending=True, inplace=True)
df_income.drop('Unknown', inplace=True)

# manipulate df_marital
df_marital = df_marital[['MaritalStatus', 'Total']].groupby(['MaritalStatus'], as_index=False).sum()
df_marital.set_index('MaritalStatus', inplace=True)
df_marital.sort_values(by='Total', ascending=True, inplace=True)
df_marital.drop('Unknown', inplace=True)

# manipulate df_age
df_age = df_age[['CustomerAge', 'Total']].groupby(['CustomerAge'], as_index=False).sum()
df_age.set_index('CustomerAge', inplace=True)
df_age.sort_values(by='Total', ascending=True, inplace=True)
```
